```python
import pyLDAvis
import pyLDAvis.gensim
import pickle
from gensim.models.ldamulticore import LdaMulticore
```

```python
dictionary = pickle.load(open('dictionary_and_corpus/M/dictionary', "rb"))
corpus_tfidf = pickle.load(open('dictionary_and_corpus/M/corpus_tfidf', "rb"))
lda_model = LdaMulticore.load('topic_models/M/model27')

vis = pyLDAvis.gensim.prepare(topic_model=lda_model, corpus=corpus_tfidf, dictionary=dictionary)
pyLDAvis.enable_notebook()
pyLDAvis.display(vis)
```
