```python
import pandas as pd
import ast
from string import digits
import re
from gensim.models.ldamulticore import LdaMulticore
import pickle
dictionary = pickle.load(open('dictionary_and_corpus/DGWS/dictionary', "rb"))
corpus_tfidf = pickle.load(open('dictionary_and_corpus/DGWS/tfidf_corpus', "rb"))
def get_words(words_of_topic):
    temp = words_of_topic.replace("\"",'')
    temp = temp.replace('*','')
    temp = temp.replace('.','')
    remove_digits = str.maketrans('', '', digits)
    temp = temp.translate(remove_digits)
    temp = temp.split('+')
    return temp

def scores_of_word(words_of_topic):
    temp = words_of_topic.replace("\"",'')
    temp = temp.replace('*','')
    temp =re.findall("\d+\.\d+", temp)
    return temp

def words_and_scores(words_of_topic):
    temp = words_of_topic.replace("\"",'')
    temp = temp.replace('*',' ')
    temp = temp.split('+')
    return temp

def get_dominant_topic(list: doc):
    vec_bow = dictionary.doc2bow(doc)
    topic_distribution = dict(lda_model[vec_bow])
    dominant_topic = max(topic_distribution, key=topic_distribution.get)
    return dominant_topic
```

```python
f = open('topic_models/M/topics.txt')
string_topics = f.read()
topics = ast.literal_eval(string_topics)
data = {}
for i in topics:
    data['topic ' + str(i[0])] = words_and_scores(i[1])
pd.DataFrame(data=data)
```

```python
lda_model = LdaMulticore.load('topic_models/M/model27')
topics = lda_model.print_topics(num_topics=30)
data = {}
for i in topics:
    data['topic ' + str(i[0])] = get_words(i[1])
pd.DataFrame(data=data).T
```
